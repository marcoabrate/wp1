### Sample Static Filters

```python
df_sg = df[id_cols+sg_cols].dropna()

len(df_sg)
```

```python
df_sg[df_sg.pref_phase_sg==0]
```

```python
ori = 60
sf = 0.02
phase = 0

g = cv2.getGaborKernel(
    ksize=(80,80),
    sigma=1/sf, # 50
    theta=ori*deg_to_rad,
    lambd=1/sf,
    gamma=1, # 1
    psi=phase*2*np.pi, 
)

plt.figure(figsize=(5, 4))
plt.imshow(g, cmap='gray')
plt.show()
```

```python
im_conv = cv2.filter2D(
    im_gray,
    ddepth=cv2.CV_32F,
    kernel=cv2.flip(g,-1),
    borderType=cv2.BORDER_CONSTANT
)
```

```python
def convolution2d(image, kernel, padding=None):
    kernel = cv2.flip(kernel,-1)
    m, _ = kernel.shape
    
    if padding is not None:
        p = int(np.floor(m/2))
        image_pad = cv2.copyMakeBorder(
            image,
            top=p,
            bottom=p,
            left=p,
            right=p,
            borderType=cv2.BORDER_CONSTANT,
            value=padding,
        )
        print(f"Image dimensions changed from {image.shape} to {image_pad.shape}")
    else:
        image_pad = image
    
    y, x = image_pad.shape
    y = y - m + 1
    x = x - m + 1
    new_image = np.zeros((y,x))
    for i in range(y):
        for j in range(x):
            new_image[i][j] = np.sum(image_pad[i:i+m, j:j+m]*kernel)
    return new_image

im_conv_manual = convolution2d(im_gray, g, padding=0)
```

```python
im_conv_clip = im_conv.copy()
im_conv_clip[im_conv_clip<0] = 0
im_conv_clip[im_conv_clip>255] = 255
```

```python
im_conv_scipy = scipy.ndimage.convolve(
    (im_gray - im_gray.min()) / (im_gray.max() - im_gray.min()),
    g,
    mode='constant',
)
```

```python
im_conv_norm = (255*(im_conv - im_conv.min())/(im_conv.max() - im_conv.min())).astype(np.uint8)
im_conv_manual_norm = (255*(im_conv_manual - im_conv_manual.min())/(im_conv_manual.max() - im_conv_manual.min())).astype(np.uint8)
im_conv_scipy_norm = (255*(im_conv_scipy - im_conv_scipy.min())/(im_conv_scipy.max() - im_conv_scipy.min())).astype(np.uint8)
```

```python
print(np.sum(im_conv_norm != im_conv_manual_norm))
print(np.sum(im_conv_norm != im_conv_scipy_norm))
print(np.sum(im_conv_manual_norm != im_conv_scipy_norm))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 15))
ax[0,0].axis('off')
ax[0,0].set_title("opencv convolution")
ax[0,0].imshow(im_conv, cmap='gray')

ax[0,1].axis('off')
ax[0,1].set_title("convolution scipy")
ax[0,1].imshow(im_conv_scipy, cmap='gray')

ax[1,0].axis('off')
ax[1,0].set_title("manual convolution")
ax[1,0].imshow(im_conv_manual, cmap='gray')

ax[1,1].axis('off')
ax[1,1].set_title("convolution clipped")
ax[1,1].imshow(im_conv_clip, cmap='gray')

plt.show()
```

```python
for _, r in df_sg.sample(15).iterrows():
    ori = r['pref_ori_sg']
    sf = r['pref_sf_sg']
    phase = r['pref_phase_sg']
    
    g = cv2.getGaborKernel(
        ksize=(80,80),
        sigma=1/sf, # 50
        theta=ori*deg_to_rad,
        lambd=1/sf,
        gamma=1, # 1
        psi=phase*2*np.pi, 
    )
    
    im_conv = cv2.filter2D(
        im_gray,
        ddepth=cv2.CV_32F,
        kernel=cv2.flip(g,-1),
        borderType=cv2.BORDER_CONSTANT
    )

    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    ax[0].imshow(im_conv, cmap='gray')
    ax[0].axis('off')
    ax[1].imshow(g, cmap='gray')
    ax[1].axis('off')
    
    plt.title(f"{ori}, {sf}, {phase}")
    plt.axis('off')
    plt.show()
```
